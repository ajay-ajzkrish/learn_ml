```python
import numpy as np
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('db_data.csv')
df = df.drop('out',1)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2, random_state=100)
model.fit(df)
df['km_label'] = model.labels_
```

```python
plt.scatter(df['F1'], df['F2'],c = df['km_label'])
plt.show()
```

```python
from sklearn.cluster import DBSCAN

df = pd.read_csv('db_data.csv')
df = df.drop('out',1)

db = DBSCAN(eps = 0.2, min_samples=5)
db.fit(df)
df['db_label' ] = db.labels_

plt.scatter(df['F1'], df['F2'],c = df['db_label'])
plt.show()
```

```python
from sklearn.cluster import DBSCAN

df = pd.read_csv('db_data.csv')
df = df.drop('out',1)

db = DBSCAN(eps = 0.1, min_samples=5)
db.fit(df)
df['db_label' ] = db.labels_

plt.scatter(df['F1'], df['F2'],c = df['db_label'],cmap='viridis')
plt.show()
```
