```python
import seaborn as sns
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

from pylab import rcParams

import warnings
warnings.filterwarnings("ignore")
```

```python
rcParams['figure.figsize'] = [10,6]
```

```python
df = pd.read_csv('./Marketing/Marketing 2.0.csv')
df.head()
```

```python
df.rename(columns={'Social Media':'Social_Media'}, inplace = True)
```

```python
cols = list(df.drop('Sales', axis = 1).columns)
```

```python
for i in cols:
    indices = cols.index(i)
    
    plt.subplot(1,3,indices + 1)
    sns.scatterplot(x = i, y = 'Sales', data = df)
    plt.title(i)
    
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(df.corr(), annot = True,cmap = 'plasma')
plt.show()
```

```python
import statsmodels.formula.api as sfa

model_tv = sfa.ols('Sales~TV', data = df).fit()
model_tv.summary()
```

```python
model_tv = sfa.ols('Sales~Radio', data = df).fit()
model_tv.summary()
```

```python
model_tv = sfa.ols('Sales~Social_Media', data = df).fit()
model_tv.summary()
```

```python
model_tv = sfa.ols('Sales~TV+Social_Media+Radio', data = df).fit()
model_tv.summary()
```
