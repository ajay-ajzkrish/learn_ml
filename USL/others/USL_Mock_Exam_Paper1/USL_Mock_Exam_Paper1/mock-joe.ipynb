```python
import numpy as np
import pandas as pd
```

```python


# Generate left-skewed data with 500 points
data = np.random.exponential(scale=5, size=500) + 1

# Sort the data in ascending order to make it left-skewed
right = pd.DataFrame(data,columns = ['asc'])
left = pd.DataFrame(-data + np.max(data) + 1,columns = ['desc'])

# You can print the first 10 values as an example
# print(data[:10])
```

```python
import seaborn as sns
```

```python
sns.displot(right)
```

```python
sns.displot(left)
```

```python
left_log = np.log(left)
right_log = np.log(right)
```

```python
sns.displot(left_log)
```

```python
sns.displot(right_log)
```
