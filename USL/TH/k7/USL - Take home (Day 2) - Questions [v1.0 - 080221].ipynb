```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agc = AgglomerativeClustering(n_clusters=2,linkage='ward')
agc.fit(df_sc)
```

```python
data['labels'] = agc.labels_
```

```python
data.head()
```

```python
data['labels'].value_counts()
```

```python
#From the above code we can infer that cluster 2 has 40 datapoints and cluster 1 has 21 datapoints
```
